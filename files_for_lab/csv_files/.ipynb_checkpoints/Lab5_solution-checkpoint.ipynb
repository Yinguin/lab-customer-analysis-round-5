```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

from sklearn.preprocessing import Normalizer
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error

import warnings
warnings.filterwarnings('ignore')
```

### Get Data

```python
data = pd.read_csv('marketing_customer_analysis.csv')
data
```

### Dealing with Data

```python
def clean_data(df):
    df.columns=[col_name.lower().replace(' ', '_') for col_name in df.columns]
    df=df.drop(df.columns[0], axis=1)
    df=df.dropna()
    df['effective_to_date']=pd.to_datetime(df['effective_to_date'], errors='coerce')
    return df

clean_data(data)
```

### Explore Data

```python
data.describe()
```

```python
data['response'].value_counts().plot.bar()
plt.show()
```

```python
data_response = data
data_response['response'] = data_response['response'].replace({'Yes': 1, 'No': 0})
data_response['response'].value_counts()
```

```python
data1 = data_response.groupby(['sales_channel'])[['response']].sum()
data1['response_rate_by_sales_channel'] = data1['response'] / data_response['sales_channel'].value_counts()
sns.barplot(x=data1.index.values, y='response_rate_by_sales_channel', data=data1)
plt.xlabel('Sales Channel')
plt.title('Response Rate by Sales Channel')
plt.show()
```

```python
data2 = data_response[['total_claim_amount','response']].sort_values('total_claim_amount')
data2
```

```python
data2_group = data2.groupby(pd.cut(data2['total_claim_amount'], 
                             bins=[0, 1000, 2000, 3000],  
                             labels=['low (0-1000)', 'medium (1000-2000)','high (2000-3000)']))['response'].agg(['size', 'sum']).reset_index()
data2_group['response_rate_by_total_claim_amount'] = data2_group['sum'] / data2_group['size']
data2_group
```

```python
sns.barplot(x='total_claim_amount', y='response_rate_by_total_claim_amount', data=data2_group)
plt.xlabel('Total Claim Amount')
plt.title('Response Rate by Total Claim Amount')
plt.show()
```

```python
data3 = data_response[['income','response']].sort_values('income')
data3
```

```python
data3_group = data3.groupby(pd.cut(data3['income'], 
                             bins=[0, 20000, 40000, 60000, 80000, 100000]))['response'].agg(['size', 'sum']).reset_index()
data3_group['response_rate_by_income'] = data3_group['sum'] / data3_group['size']
data3_group
```

```python
plt.figure(figsize=(10,8))
sns.barplot(x='income', y='response_rate_by_income', data=data3_group)
plt.xlabel('Income')
plt.title('Response Rate by Income')
plt.show()
```
